```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

df=pd.read_csv('ecg_new.csv')
frame=df.drop(['time(sec)'],axis=1)
new=frame.head(1000)
plt.show(sns.regplot(x=df['time(sec)'],y=df['ch1'],data=new,x_jitter=0.5,y_jitter=0.5,x_estimator=np.mean))
#print(new)
```
